```python
# Polynomial Regression
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
page_speed = np.random.normal(3,1,1000)
purchase_amount = np.random.normal(50,10,1000) / page_speed
```

```python
sns.scatterplot(x=page_speed,y=purchase_amount)
```

```python
coeff = np.polyfit(page_speed,purchase_amount,8)
```

```python
coeff
```

```python
poly = np.poly1d(coeff)
```

```python
print(poly)
# np.polyfit(...) fits a polynomial of degree 8 to the data. The output is an array of coefficients.

# np.poly1d(coeff) creates a polynomial function from those coefficients.

# So now poly(x) represents your polynomial regression model.
```

```python
x_line  = np.linspace(min(page_speed), max(page_speed), 100)
```

```python
y_line = poly(x_line)
# y_line is the set of predicted y-values for a smooth curve using the polynomial model
# over a range of x-values.
```

```python
y_pred = poly(page_speed)
```

```python
plt.scatter(page_speed, purchase_amount, color='red', label='Data points')  # Red scatter plot of actual points
plt.plot(x_line, y_line, label='Polynomial Fit (deg=8)')  # Curve from polyfit
plt.legend()
plt.title("Polynomial Regression")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import r2_score, mean_squared_error
print("R:", r2_score(purchase_amount, y_pred))
print("MSE:", mean_squared_error(purchase_amount, y_pred))
```
