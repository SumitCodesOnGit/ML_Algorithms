# Reinforcement Learning: Cart Wheel Balance
Objective: To learn and understand Reinforcement Learning

Reinforcement Learning is a type of Machine Learning where an agent learns to make decisions
by interacting with an environment to achieve a goal.

1. Agent - learner or decision maker
2. Environment - everything outside the agent that it interacts with.
3. State (S) - a snapshot of the environment at a given time.
4. Action (A) - all possible moves an agent can make.
5. Reward  (R) - Feedback signal to evaluate the action's effectiveness.
6. Policy - strategy used by an agent to choose actions based on state.
7. Value Function (V(s)) - expected return from state s
8. Q-Value (Q(s,a)) - expected return for taking action a in state s.
9. Episode - a full sequence from start to terminal state.
10. Trajectory - sequence of (state, action, reward) tuples.

Types of problems in RL:
1. Model-Free RL - agent learns through trial and error.
2. Madel-Based RL - agent tries to build a model of the evironment.
3. Exploration vs Exploitation
   Exploration - Try new actions to gather more info.
   Exoloitation - Use current knowledge to maximize reward.

Important Algorithms:
1. Value Based Methods
2. Policy Based Method
3. Model Based Method
4. Deep RL

```python
import gymnasium as gym
import numpy as np
```

```python
env = gym.make('CartPole-v1', render_mode='human')
```

```python
obs, info = env.reset()
```

```python
for step in range(1000):
    action = env.action_space.sample()
    obs, reward, terminated, truncated, info = env.step(action)
    print(f"Step: {step}, Action: {action}, Reward: {reward}, obs= {obs}")
    if terminated or truncated:
        obs, info = env.reset()

env.close()
    
```
